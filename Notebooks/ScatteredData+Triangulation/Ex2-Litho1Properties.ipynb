```python
## Appears to be broken !!


# def profile0(model, ll0, ll1, resolution, samples, QID):
#     """
#         model: self
#         ll0: (lon0, lat0)
#         ll1: (lon1, lat1)
#         resolution: separation (km) of locations to sample along the profile
#         samples: number of depths to sample vertically in the model
#     """
    
#     import pyproj

#     startlong = ll0[0]
#     startlat  = ll0[1]
#     endlong   = ll1[0]
#     endlat    = ll1[1]

#     g = pyproj.Geod(ellps='WGS84')

#     (az12, az21, dist) = g.inv(startlong, startlat, endlong, endlat)

#     # calculate line string along path with segments <= 1 km
    
#     lonlats = np.array(g.npts(startlong, startlat, endlong, endlat, 1 + int(dist / (1000.0 * resolution))))
#     data = np.empty( (lonlats.shape[0], samples, 2))
    
#     upper = model.layer_depth(lonlats.T[1], lonlats.T[0], "ICE-TOP") * 1.0e-3
#     lower = model.layer_depth(lonlats.T[1], lonlats.T[0], "LID-BOTTOM") * 1.0e-3

#     for i in range(0,len(upper)):
#         data[i,:,0] = np.linspace(lower[i],upper[i], samples)
    
#     for s in range(0, samples):
#         depths = data[:,s,0]        
#         c, profile = model.property_at_location(lonlats.T[1], lonlats.T[0], depths, QID)       
#         data[:,s,1] = profile[:]
    
#     return lonlats, data
    
    
    
def great_circle_profile(ll0, ll1, depths, resolution, QID):
    """
        model: self
        ll0: (lon0, lat0)
        ll1: (lon1, lat1)
        depths (km)
        resolution: separation (km) of locations to sample along the profile
    """
    
    import pyproj

    startlong = ll0[0]
    startlat  = ll0[1]
    endlong   = ll1[0]
    endlat    = ll1[1]

    g = pyproj.Geod(ellps='WGS84')

    (az12, az21, dist) = g.inv(startlong, startlat, endlong, endlat)

    # calculate line string along path with segments <= 1 km
    
    lonlats = np.array(g.npts(startlong, startlat, endlong, endlat, 1 + int(dist / (1000.0 * resolution))))
    data = np.empty( (lonlats.shape[0], depths.shape[0]))
    
    
    for s,ll in enumerate(lonlats):
        
        c, profile = litho.property_on_profile(ll[1], ll[0], depths, QID)   
        data[s,:] = profile[:]
    
    
    return lonlats, data
    
    
    
    
```
